```python
def replace_inf_nan(lst):
    # Convert list to a numpy array for vectorized operations
    arr = np.array(lst)

    # Create a mask where inf, nan, or -inf values are True
    mask = np.isnan(arr) | np.isinf(arr) | np.isneginf(arr)

    # Replace these elements with 1
    arr[mask] = 0

    return arr.tolist()  # Convert back to a list if needed
```

```python
# the Jean curve
import time


N = 20 # physical N
J = 100
beta1 = 0.91

t1 = 2

grange = np.linspace(0,10,300)

PJ_initial = [BurnProb(N,J,beta1,g) for g in tqdm(grange)]

s1 = time.time()
dist1 = [P(N-1,n,t1,beta1,0) for n in range(J)]
absorb1 = absorbProb(t1,N,beta1,0)
s2 = time.time()
print("total:", absorb1 + sum(dist1))
print('time',s2-s1)
# plt.plot(grange,PJ_initial)
```

```python
PJ_later2 = []
PJ_later3 = []
beta2 = 1.2
beta3 = 1.5
t2 = 1
for g in tqdm(grange):
    # compute population distribution taking into account the constant suppression at lower birth rate
    dist1 = np.asarray([P(N-1,n,t1,beta1,g) for n in range(J)])
    
    # propagate for a wind event of a short duration, increasing the birth rate, with the constant suppression
#     dist2 = np.asarray([sum([dist1[n]*P(n,m,t2,beta2,g) for n in range(J)]) for m in range(J)])
    
    # use footprint dist heuristic 
    ftpt1 = [max(N-1,n)+1 for n in range(J)]
    
    # regarding dist1 as an ensemble, compute the burn probability for each initial value with lower burn probability 
    
    individual_PJs = np.asarray([BurnProb(i+1,J-ftpt1[i],beta2,g) for i in range(J)])
    product = dist1*individual_PJs
    PJ_later2.append(sum(product))
    
    individual_PJs = np.asarray([BurnProb(i+1,J-ftpt1[i],beta3,g) for i in range(J)])
    product = dist1*individual_PJs
    PJ_later3.append(sum(product))
```
